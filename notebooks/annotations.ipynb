Reading evaluation dataset produced by GREEN model

```python
PROMPT_VERSIONS = [1, 2]
REGIONS = ['bone', 'heart', 'lung', 'mediastinal']

df_list = []
for prompt_version in PROMPT_VERSIONS:
    for region in REGIONS:
        res_df = pd.read_csv(os.path.join(DATA_PATH, 'reports_annotations_train_eval_{prompt_version}_{region}.csv'.format(prompt_version=prompt_version, region=region)))
        res_df['region'] = region
        res_df['prompt_version'] = prompt_version
        df_list.append(res_df)

res_df = pd.concat(df_list, axis=0)

results = res_df.groupby(['prompt_version','region'])['green_score'].mean()

results
```

```python
res_df.groupby(['prompt_version'])['green_score'].mean()
```

```python
df_list = []
for prompt_version in PROMPT_VERSIONS:
    for region in REGIONS:
        res_df = pd.read_csv(os.path.join(DATA_PATH, 'reports_annotations_test_eval_{prompt_version}_{region}.csv'.format(prompt_version=prompt_version, region=region)))
        res_df['region'] = region
        res_df['prompt_version'] = prompt_version
        df_list.append(res_df)

res_df = pd.concat(df_list, axis=0)

results = res_df.groupby(['prompt_version','region'])['green_score'].mean()

results
```
